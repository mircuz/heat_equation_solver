```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("thermal_results.dat",'r') as f:
    f.seek(0)
    txt=f.readlines()
```

```python
#Fill the array
dim_x = int(txt[1])
dim_y = int(txt[2])
dim_t = int((len(txt)-4))
array = np.zeros((dim_x,dim_y,dim_t//dim_x),dtype=float)
x=y=t=0
for line in txt[4:dim_t]: 
    
    line.strip()
    row=line.split()
    
    for y in range(dim_y):
        array[x,y,t] = float(row[y])
        #print(x,y,t,array[x][y][t])
       
    x =x+1
    if (x>=dim_x) :
        x=0
        t=t+1
```

```python
#Plot
for time in range(dim_t//dim_x -1):
    img = (array[:,:,time])
    plt.imshow(img);
    plt.pause(0.1);
    plt.draw();
```
